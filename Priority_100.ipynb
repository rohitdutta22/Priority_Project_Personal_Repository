### One-Hot Encoding Categorical Variables

```python
data = pd.get_dummies(data, columns = ['Frequency.of.Customer'])
```

```python
list(data.columns)
```

```python
data.head()
```

```python
data['int_freq_high_no_of_items'] = data['No..of.Items.requested'] * data['Frequency.of.Customer_High']
data['int_freq_low_no_of_items'] = data['No..of.Items.requested'] * data['Frequency.of.Customer_Low']
data['int_freq_medium_no_of_items'] = data['No..of.Items.requested'] * data['Frequency.of.Customer_Medium']
data['int_freq_new_no_of_items'] = data['No..of.Items.requested'] * data['Frequency.of.Customer_New']
```

```python
data.head()
```

```python
data['int_freq_high_per_ordval'] = data['Per_order_value'] * data['Frequency.of.Customer_High']
data['int_freq_low_per_ordval'] = data['Per_order_value'] * data['Frequency.of.Customer_Low']
data['int_freq_medium_per_ordval'] = data['Per_order_value'] * data['Frequency.of.Customer_Medium']
data['int_freq_new_per_ordval'] = data['Per_order_value'] * data['Frequency.of.Customer_New']
```

```python
data.head(10)
```

```python
data = data.drop(['Customer.Country'], axis = 1)
```

### Train-Test Split

```python
X = data.drop(['Priority.Score..out.of.1.10.'], axis = 1)
y = data['Priority.Score..out.of.1.10.']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state = 7)
```

```python
x_test.shape
```

```python
y_train.value_counts()
```

### Handling Imbalance in Data

###### SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy = 'minority')
X_sm, y_sm = smote.fit_resample(x_train, y_train)
```

```python
y_sm.value_counts()
```

```python
smote1 = SMOTE(sampling_strategy = 'minority')
X_sm1, y_sm1 = smote.fit_resample(X_sm,y_sm)
```

```python
y_sm1.value_counts()
```

```python
smote2 = SMOTE(sampling_strategy = 'minority')
X_sm2, y_sm2 = smote.fit_resample(X_sm1,y_sm1)
```

```python
y_sm2.value_counts()
```

```python
smote3 = SMOTE(sampling_strategy = 'minority')
X_sm3, y_sm3 = smote.fit_resample(X_sm2,y_sm2)
```

```python
y_sm3.value_counts()
```

```python
X_sm3.shape
```

## ANN

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y1 = encoder.fit_transform(y_sm3)
#Y = pd.get_dummies(y1).values
```

```python
model = keras.Sequential([
    keras.layers.Dense(32, input_shape=(27,), activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    #keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(5, activation='softmax')
])
```

```python
model.compile(optimizer='Adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_sm3, y1, batch_size = 50, epochs=200)
```

```python
y_pred = model.predict(x_test)
```
